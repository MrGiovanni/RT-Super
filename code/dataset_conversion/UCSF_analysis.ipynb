```python
import pandas as pd
```

```python
meta = pd.read_csv('/Users/pedrobassi/Downloads/UCSF_metadata_filled.csv')
```

```python
meta
```

```python
#columns with the word number of lesion
lesion_cols = [col for col in meta.columns if 'lesion instances' in col]
```

```python
lesion_cols
```

```python
#print sum for each column
for col in lesion_cols:
    print(col, meta[col].sum())
```

```python
chosen_tumors_new = ['colon','adrenal gland', 'esophagus', 'spleen', 'uterus', 'bladder', 'gallbladder', 'prostate', 'duodenum']
```

```python
chosen_tumor_columns = [f'number of {tumor} lesion instances' for tumor in chosen_tumors_new]
```

```python
#get all the rows that have any of the chosen_tumor_columns >=1 and rows where all tumors are 0
chosen_cases_pos = meta[(meta[chosen_tumor_columns].sum(axis=1) >= 1)]
chosen_cases_neg = meta[(meta[lesion_cols].sum(axis=1) == 0)]
```

```python
chosen_cases_pos
```

```python
chosen_cases_neg
```

```python
#mix both
chosen_cases = pd.concat([chosen_cases_pos, chosen_cases_neg])
```

```python
# cases we can use (no unk size)
usable = pd.read_csv('/Users/pedrobassi/Downloads/UCSFLLMOutputLarge27k.csv')
```

```python
chosen_cases = chosen_cases[chosen_cases['BDMAP ID'].isin(usable['BDMAP_ID'])]
```

```python
chosen_cases
```

```python
#remove cases w/o BDMAP ID
chosen_cases = chosen_cases.dropna(subset=['BDMAP ID'])
```

```python
chosen_cases
```

```python
#save to csv
chosen_cases.to_csv('/Users/pedrobassi/Downloads/UCSF_multi_tumor_colon_adrenalgland_esophagus_spleen_uterus_bladder_gallbladder_prostate_duodenum_normal.csv', index=False)
```
